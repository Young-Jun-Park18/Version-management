# 1. Branch 사용

a. 작업을 하다가 master 브랜치를 건들지 않고 여러 테스트를 하고 싶을 때 사용합니다.

b. 새 브랜치 이름을 작성하고 생성합니다

c. 새로 만든 브랜치가 잘 선택되었는지 확인하고 코드 작업을 시작합니다.

d. commit 후 push 합니다

### 새로운 브랜치에서 작업할 때 절대 master 브랜치에 push하면 안 됨

# 2. branch 복구

a. 브랜치 생성 후 작업하다가 이전으로 돌아가고 싶을 때 사용합니다.

만약 마지막 줄을 작성하지 않은 곳으로 돌아가고 싶다면 돌아가고 싶은 곳의 히스토리를 선택합니다

이 부분이 돌아가야할 부분입니다.

우클릭 후 "이 커밋까지 현재 브랜치를 초기화"를 누릅니다

hard를 선택하면 커밋도 취소시키고 현재 작업중인 vscode의 코드까지 해당 위치로 돌아가게합니다

제일 마지막 줄이 사라진 것을 확인합니다.

# 2-1 branch 복구 후 pull 문제 해결

hard 복구를 하게 되면 git-hub repo에 올라간 코드와 현재 작업물이 불일치하여 pull에 빨간불이 들어옵니다 
하지만 저걸 누르면 복구 이전으로 다시 돌아갑니다

터미널을 열어 강제 푸쉬를 해줘야합니다

### "git push 저장소이름 브랜치이름 --force"

강제 업데이트를 마쳤다면 pull 을 통해 형식적으로 빨간불은 없애줍니다

# 3. brach 병합

브랜치에서 테스트한 모든 코드가 정상 작동하고 master branch에 적용해도 되겠다 싶으면 병합을 합니다

# 3-1 master 브랜치를 선택합니다

현재 위치가 master 임을 확인하고 병합할 branch를 우클릭합니다

이제 test2 라는 브랜치에서만 코드 변경이 일어나는 것이 아니라 master라는 메인 branch에 test2에서 이것저것 추가해본 코드가 병합되게 됩니다

# 4. branch 삭제

병합을 마치고 테스트가 끝난 branch는 삭제해도 됩니다
터미널을 열고 명령어를 입력합니다

### git push 저장소이름 --delete 삭제할브랜치이름


저는 깃허브 저장소 이름이 fifth-dimension-develop이고, 삭제할 브랜치는 test와 test2입니다

###  " git push fifth-dimension-develop --delete test2 " 입력

깨끗하게 삭제된 것을 확인할 수 있습니다

source tree local에서도 삭제하기 위해 해당 브랜치 우클릭 후 삭제를 누릅니다
