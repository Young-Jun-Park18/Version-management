# 2. build error가 나는 이유

2-1 나는 레포지토리에 dna라는 폴더를 만들어놨기 때문에 Base directory 입력칸에 dna를 입력했다 (폴더 없으면 빈칸으로 놔둬야함)

2-2 나는 react + vite를 사용하기 떄문에 build command 입력칸에 npm run build를 입력했다 (이건 빌드툴마다 명령어 다르니까 확인)

2-3 publish directory에는 dist를 입력해준다 (vite를 사용하지 않는 사람은 명령어가 다르므로 확인바람)

2-4 마지막으로 배포전에 build 파일을 만들어놔야하는데 이 작업을 하지 않았다면 새 터미널창을 열어서 npm run build를 하자
