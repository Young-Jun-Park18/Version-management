## 5. Routes와 Route에 대하여

### 5-1 Route는 각각의 페이지라고 생각하면 됩니다. Route 태그 안에 path =  " ~경로 " 속성으로 경로를 설정할 수 있습니다. 

슬래쉬는 현재 메인 화면 경로를 의미하고, /Members 는 개발자 마음대로 작명할 수 있는 부분입니다. 

element = {} 속성에는 해당 페이지의 내용을 의미합니다. html 코드인 div 태그를 넣거나, 이미 짜여진 페이지를 컴포넌트로 넣는 것은 개발자의 선택입니다.

제 코드에는 Members라는 경로에 "멤버 소개 공간입니다." 라는 div 태그를 넣었으니 해당 경로로 이동하면 메세지가 뜨는 것을 확인할 수 있습니다. 

### 5-2 사용자가 직접 경로를 지정하여 접속하지 않으니 Link 태그로 경로를 설정할 수 있습니다.

저는 홈페이지의 Member라는 리스트를 클릭하면 /Members 로 경로가 변경되도록 설정했습니다.

Member를 사용자가 클릭하면 주소가 변경되면서 해당 페이지의 내용이 출력됩니다.
