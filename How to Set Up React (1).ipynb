# 4-1 에러발생시

1. 나는 바탕화면에 React라는 폴더를 만들었고
2. 그 폴더 안에 blog라는 폴더가 있고
3. blog 안에 package.json 파일이 들어있다
4. 하지만 에러 3번째 줄처럼 vscode는 React 폴더 안에서 package.json 파일을 찾고 있다.
5. 따라서 경로를 React 안에 blog까지 들어갈 수 있도록 설정해줘야한다

1. pwd를 입력해서 현재 경로를 확실하게 확인하고

2. 앞에 cd 명령어를 통해 경로를 blog까지 재설정 해준다
3. 다시 pwd로 현재 경로를 확인한다

4. 이후 다시 npm run dev를 입력한다

5. 나오는 링크를 크롬에 복붙하면 성공


# 5. 각종 파일들의 기능

1. public 폴더는 쓸데없는 이미지 파일이나 데이터를 보관하는 곳
2. app.jsx는 내가 실제로 작업하는 공간
3. package.json 파일은 내가 설치한 라이브러리를 기록해주는 파일
4. 동작 과정은 app.jsx를 main.jsx로 보내고 index.html에 보내는 형식
5. node_modules는 설치한 라이브러리 전부 보관(열어볼 일은 별로 없음)

# 6. App.css와 index.css 파일 내용은 전부 지우고 시작하는 것이 작업하기 편함

# A. 위 방식과는 다른 방법

1. 빈폴더를 만들고 관리자 권한으로 vscode를 켜서 해당 폴더를 오픈한다
2. 터미널을 열고 "npx create-react-app 폴더이름" 을 따옴표 뺴고 작성한다.
3. 설치를 완료했다면 다시 file -> 폴더열기 -> 방금 만든 폴더이름을 선택하고 연다.
4. 다시 터미널에 npm start를 입력한다.
5. 에러가 날 경우에는 pwd로 현재 폴더를 확인하고 cd 를 사용하여 가장 안쪽 폴더로 이동시켜준다
6. npm start로 미리보기 가능

# B. npm start error 

1. npm start 명령어가 error날 경우에는 다음을 고려해보자
2. 터미널에 npm install react-scripts 를 입력하고 설치해보기
3. 터미널에 npm install 입력하고 설치해보기 
4. 터미널에 npm install web-vitals 설치해보기

# C.주의할 점 :  "npx create-react-app 폴더이름" 으로 폴더를 만들었으면 항상 다시 file -> 폴더열기 -> 방금 만든 폴더를 선택해야 에러를 80%이상 줄일 수 있음

# D. 보통 옛날 버전에는 .js로 나오고 최신 버전에는 .jsx로 나옴

1. 실행 명령어도 옛날 버전은 npm start이고
2. 최신 버전은 npm run dev임
